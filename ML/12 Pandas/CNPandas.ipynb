# Handling String data in table

```python
# Let we have a column with name Gender in that we have two entries 'Male' and 'Female' and now we want to do that all Male became 1 and female 0 ok..
dataK['Gender']='Male'
dataK.head(20)
```

```python
# making few entries of female
dataK.iloc[0:5, 11]= "Female"
dataK.head(20)
```

```python
# now we made new column "sex" in which we made all the entry of male as 1 and female as 0
def fxn(str):
    if str=="Male":
        return 1
    else:
        return 0
dataK['sex']= dataK.Gender.apply(fxn)
dataK.head(20)
```

```python
# now i am dropping the gender column and work on the sex column ok..
del dataK['Gender']
dataK.head(20)
```
